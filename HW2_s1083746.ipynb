## Read data from csv file

```python
import pandas as pd
```

```python
#display the information of the data
df = pd.read_csv('COVID-19_data.csv', parse_dates = ['dateRep'], dayfirst=True)
print(df.dtypes)
print(df.shape)
display(df.head())
```

```python
#set and sort "dateRep"
df = df.set_index('dateRep')
df = df.sort_index()

#Take sorting data from collumns based on "countryterritoryCode"
df1 = df.loc[df['countryterritoryCode']=='USA']
display(df1.head())

df2 = df.loc[df['countryterritoryCode']=='ITA']
display(df2.head())

df3 = df.loc[df['countryterritoryCode']=='JPN']
display(df3.head())
```

# Plotting data

```python
import matplotlib.pyplot as plt
```

```python
#Plotting the data and detect their information
plt.figure(figsize=(10, 3), dpi=200, facecolor='w', edgecolor='k')
plt.plot(df1.index, df1['cases'], label='USA')
plt.plot(df2.index, df2['cases'], label='ITA')
plt.plot(df3.index, df3['cases'], label='JPN')

#Adding information for the graph
plt.xlabel('Time')
plt.ylabel('Cases')
plt.title('Covid 19 in USA, Japan and Italy')

#Adding legend to show label table
plt.legend()

#Adding grid to the graph
plt.grid(True)
```
